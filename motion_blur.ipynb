```python
import os
import cv2
import numpy as np
from multiprocessing import Pool, cpu_count
from tqdm import tqdm  # For progress bar

def apply_motion_blur(image, kernel_size=15, angle=0):
    """
    Apply motion blur to an image.
    Args:
        image: Input image as a NumPy array.
        kernel_size: Size of the motion blur kernel.
        angle: Angle of the motion blur in degrees (0 for horizontal, 90 for vertical).
    Returns:
        Blurred image as a NumPy array.
    """
    # Create motion blur kernel
    kernel = np.zeros((kernel_size, kernel_size), dtype=np.float32)
    kernel[int((kernel_size - 1) / 2), :] = np.ones(kernel_size, dtype=np.float32)
    kernel = cv2.warpAffine(
        kernel,
        cv2.getRotationMatrix2D((kernel_size / 2 - 0.5, kernel_size / 2 - 0.5), angle, 1),
        (kernel_size, kernel_size)
    )
    kernel /= kernel.sum()

    # Apply the kernel to the image
    blurred_image = cv2.filter2D(image, -1, kernel)
    return blurred_image

def process_image(args):
    """
    Process a single image by applying motion blur and saving it.
    Args:
        args: A tuple containing image_path, output_path, kernel_size, and angle.
    """
    image_path, output_path, kernel_size, angle = args
    try:
        image = cv2.imread(image_path)
        if image is not None:
            blurred_image = apply_motion_blur(image, kernel_size, angle)
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            cv2.imwrite(output_path, blurred_image)
            return f"Saved: {output_path}"
        else:
            return f"Failed to read image: {image_path}"
    except Exception as e:
        return f"Error processing {image_path}: {e}"

def process_dataset(input_folder, output_folder, kernel_size=15, angle=0):
    """
    Process the dataset by applying motion blur to all images and saving them.
    Args:
        input_folder: Path to the original dataset folder.
        output_folder: Path to save the processed dataset.
        kernel_size: Size of the motion blur kernel.
        angle: Angle of the motion blur in degrees (0 for horizontal, 90 for vertical).
    """

    tasks = []
    for class_name in os.listdir(input_folder):
        class_path = os.path.join(input_folder, class_name)
        if os.path.isdir(class_path):
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                if image_path.lower().endswith(('.png', '.jpg', '.jpeg')):
                    output_path = os.path.join(output_folder, class_name, image_name)
                    tasks.append((image_path, output_path, kernel_size, angle))

    print(f"Total images to process: {len(tasks)}")

    # Use multiprocessing to process images in parallel with progress bar
    with Pool(processes=cpu_count()) as pool:
        for result in tqdm(pool.imap_unordered(process_image, tasks), total=len(tasks)):
            print(result)

# Example usage:
# process_dataset('path_to_input_folder', 'path_to_output_folder', kernel_size=15, angle=45)


output_dir = "/content/drive/MyDrive/Food Image Datasets/North Indian/motion_blur"

process_dataset(base_dir, output_dir, kernel_size=15, angle=45)
```
