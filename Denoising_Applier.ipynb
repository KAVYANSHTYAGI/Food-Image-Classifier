GAUSSIAN DE NOISER [ Non-Local Means Denoising ]

```python
import os
import cv2

# Paths
input_folder = base_dir  # Replace with the path to your dataset
output_folder = "/content/drive/MyDrive/Food Image Datasets/North Indian/gaussian_nlm_filtered"  # Replace with the path where you want to save denoised images

# Function to apply Non-Local Means Denoising
def apply_denoising(image):
    return cv2.fastNlMeansDenoisingColored(
        image,
        None,
        h=10,  # Strength for denoising
        templateWindowSize=7,
        searchWindowSize=21
    )

# Function to process the dataset and save denoised images
def process_and_save_images(input_dir, output_dir):
    # Walk through the input directory
    for root, dirs, files in os.walk(input_dir):
        # Ensure output folder has the same structure
        for dir_name in dirs:
            os.makedirs(os.path.join(output_dir, dir_name), exist_ok=True)

        # Process each file
        for file_name in files:
            input_path = os.path.join(root, file_name)
            relative_path = os.path.relpath(input_path, input_dir)
            output_path = os.path.join(output_dir, relative_path)

            if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                os.makedirs(os.path.dirname(output_path), exist_ok=True)

                # Read the image
                image = cv2.imread(input_path)
                if image is None:
                    print(f"Failed to read {input_path}, skipping...")
                    continue

                # Apply the denoising filter
                denoised_image = apply_denoising(image)

                # Save the denoised image
                cv2.imwrite(output_path, denoised_image)
                print(f"Processed and saved: {output_path}")
            else:
                print(f"Unsupported file format: {file_name}")

```

```python
# Run the processing
os.makedirs(output_folder, exist_ok=True)
process_and_save_images(input_folder, output_folder)
```

SALT AND PEPPER NOISE

```python
import cv2
import os

# Paths
input_folder = base_dir  # Replace with the path to your dataset
output_folder = "/content/drive/MyDrive/Food Image Datasets/North Indian/saltpepper_median_filtered"  # Replace with the path where you want to save denoised images

# Function to apply Median Filter
def apply_median_filter(image, kernel_size=3):
    return cv2.medianBlur(image, kernel_size)

# Function to process the dataset and save filtered images
def process_and_save_images(input_dir, output_dir, kernel_size=3):
    # Walk through the input directory
    for root, dirs, files in os.walk(input_dir):
        # Ensure output folder has the same structure
        for dir_name in dirs:
            os.makedirs(os.path.join(output_dir, dir_name), exist_ok=True)

        # Process each file
        for file_name in files:
            input_path = os.path.join(root, file_name)
            relative_path = os.path.relpath(input_path, input_dir)
            output_path = os.path.join(output_dir, relative_path)

            if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                os.makedirs(os.path.dirname(output_path), exist_ok=True)

                # Read the image
                image = cv2.imread(input_path)
                if image is None:
                    print(f"Failed to read {input_path}, skipping...")
                    continue

                # Apply the median filter
                filtered_image = apply_median_filter(image, kernel_size)

                # Save the filtered image
                cv2.imwrite(output_path, filtered_image)
                print(f"Processed and saved: {output_path}")
            else:
                print(f"Unsupported file format: {file_name}")


```

```python
# Example: Apply Median Filter with a 3x3 kernel
os.makedirs(output_folder, exist_ok=True)
process_and_save_images(input_folder, output_folder, kernel_size=3)
```
