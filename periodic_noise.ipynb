```python
import os
import cv2
import numpy as np
from multiprocessing import Pool, cpu_count
from tqdm import tqdm  # For progress bar

def apply_periodic_noise(image, frequency=10, amplitude=20):
    """
    Apply periodic noise to an image.
    Args:
        image: Input image as a NumPy array.
        frequency: Frequency of the sinusoidal noise.
        amplitude: Amplitude of the sinusoidal noise.
    Returns:
        Image with periodic noise as a NumPy array.
    """
    rows, cols, _ = image.shape

    # Generate sinusoidal noise
    x = np.arange(cols)
    y = amplitude * np.sin(2 * np.pi * frequency * x / cols)
    y = y.astype(np.int32)

    # Apply noise to the image
    noisy_image = image.copy()
    for i in range(rows):
        noisy_image[i, :, :] = np.roll(image[i, :, :], y[i % len(y)], axis=0)

    return noisy_image

def process_image_periodic(args):
    """
    Process a single image by applying periodic noise and saving it.
    Args:
        args: A tuple containing image_path, output_path, frequency, and amplitude.
    """
    image_path, output_path, frequency, amplitude = args
    try:
        image = cv2.imread(image_path)
        if image is not None:
            noisy_image = apply_periodic_noise(image, frequency, amplitude)
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            cv2.imwrite(output_path, noisy_image)
            return f"Saved: {output_path}"
        else:
            return f"Failed to read image: {image_path}"
    except Exception as e:
        return f"Error processing {image_path}: {e}"

def process_dataset_periodic(input_folder, output_folder, frequency=10, amplitude=30):
    """
    Process the dataset by applying periodic noise to all images and saving them.
    Args:
        input_folder: Path to the original dataset folder.
        output_folder: Path to save the processed dataset.
        frequency: Frequency of the sinusoidal noise.
        amplitude: Amplitude of the sinusoidal noise.
    """

    tasks = []
    for class_name in os.listdir(input_folder):
        class_path = os.path.join(input_folder, class_name)
        if os.path.isdir(class_path):
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                if image_path.lower().endswith(('.png', '.jpg', '.jpeg')):
                    output_path = os.path.join(output_folder, class_name, image_name)
                    tasks.append((image_path, output_path, frequency, amplitude))

    print(f"Total images to process: {len(tasks)}")

    # Use multiprocessing to process images in parallel with progress bar
    with Pool(processes=cpu_count()) as pool:
        for result in tqdm(pool.imap_unordered(process_image_periodic, tasks), total=len(tasks)):
            print(result)

# Example usage:
# process_dataset_periodic('path_to_input_folder', 'path_to_output_folder', frequency=10, amplitude=30)

output_dir = "/content/drive/MyDrive/Food Image Datasets/North Indian/periodic_noise"
process_dataset_periodic(base_dir, output_dir, frequency=10, amplitude=20)
```

```python
import os
import cv2
import numpy as np
from multiprocessing import Pool, cpu_count
from tqdm import tqdm  # For progress bar

def apply_banding_noise(image, frequency=5, amplitude=20, direction='horizontal'):
    """
    Apply banding noise to an image.
    Args:
        image: Input image as a NumPy array.
        frequency: Frequency of the sine wave (number of bands).
        amplitude: Amplitude of the sine wave (intensity of bands).
        direction: 'horizontal' or 'vertical' for the direction of bands.
    Returns:
        Noisy image as a NumPy array.
    """
    height, width, _ = image.shape
    noisy_image = image.copy().astype(np.float32)  # Use float32 for calculations

    if direction == 'horizontal':
        x = np.linspace(0, 2 * np.pi * frequency, width)
        banding_pattern = amplitude * np.sin(x)[np.newaxis, :].astype(np.float32)
        for channel in range(3):  # Apply to all color channels
            noisy_image[:, :, channel] += banding_pattern
    elif direction == 'vertical':
        y = np.linspace(0, 2 * np.pi * frequency, height)
        banding_pattern = amplitude * np.sin(y)[:, np.newaxis].astype(np.float32)
        for channel in range(3):  # Apply to all color channels
            noisy_image[:, :, channel] += banding_pattern

    noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)  # Convert back to uint8
    return noisy_image

def process_image(args):
    """
    Process a single image by adding banding noise and saving it.
    Args:
        args: A tuple containing image_path, output_path, frequency, amplitude, and direction.
    """
    image_path, output_path, frequency, amplitude, direction = args
    try:
        image = cv2.imread(image_path)
        if image is not None:
            noisy_image = apply_banding_noise(image, frequency, amplitude, direction)
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            cv2.imwrite(output_path, noisy_image)
            return f"Saved: {output_path}"
        else:
            return f"Failed to read image: {image_path}"
    except Exception as e:
        return f"Error processing {image_path}: {e}"

def process_dataset(input_folder, output_folder, frequency=5, amplitude=20, direction='horizontal'):
    """
    Process the dataset by adding banding noise to all images and saving them.
    Args:
        input_folder: Path to the original dataset folder.
        output_folder: Path to save the processed dataset.
        frequency: Frequency of the sine wave (number of bands).
        amplitude: Amplitude of the sine wave (intensity of bands).
        direction: 'horizontal' or 'vertical' for the direction of bands.
    """
    supported_extensions = ('.png', '.jpg', '.jpeg', '.jfif', '.webp')  # Updated to include .jfif and .webp

    tasks = []
    for class_name in os.listdir(input_folder):
        class_path = os.path.join(input_folder, class_name)
        if os.path.isdir(class_path):
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                if image_path.lower().endswith(supported_extensions):
                    output_path = os.path.join(output_folder, class_name, image_name)
                    tasks.append((image_path, output_path, frequency, amplitude, direction))
                else:
                    print(f"Skipping non-image file: {image_name}")

    print(f"Total images to process: {len(tasks)}")

    # Use multiprocessing to process images in parallel with progress bar
    with Pool(processes=cpu_count()) as pool:
        for result in tqdm(pool.imap_unordered(process_image, tasks), total=len(tasks)):
            print(result)

output_dir = "/content/drive/MyDrive/Food Image Datasets/North Indian/banding3"  # Replace with your output folder path
process_dataset(base_dir, output_dir, frequency=20, amplitude=25, direction='horizontal')
```
