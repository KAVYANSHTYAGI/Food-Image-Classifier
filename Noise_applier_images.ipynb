GAUSSIAN NOISE

```python
import os
import cv2
import numpy as np
from multiprocessing import Pool, cpu_count
from tqdm import tqdm  # For progress bar

def apply_gaussian_noise(image, mean=0, std=25):
    """
    Apply Gaussian noise to an image.
    Args:
        image: Input image as a NumPy array.
        mean: Mean of the Gaussian noise.
        std: Standard deviation of the Gaussian noise.
    Returns:
        Noisy image as a NumPy array.
    """
    noise = np.random.normal(mean, std, image.shape)
    noisy_image = image + noise
    noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)
    return noisy_image

def process_image(args):
    """
    Process a single image by adding Gaussian noise and saving it.
    Args:
        args: A tuple containing image_path, output_path, mean, and std.
    """
    image_path, output_path, mean, std = args
    try:
        image = cv2.imread(image_path)
        if image is not None:
            noisy_image = apply_gaussian_noise(image, mean, std)
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            cv2.imwrite(output_path, noisy_image)
            return f"Saved: {output_path}"
        else:
            return f"Failed to read image: {image_path}"
    except Exception as e:
        return f"Error processing {image_path}: {e}"

def process_dataset(input_folder, output_folder, mean=0, std=25):
    """
    Process the dataset by adding Gaussian noise to all images and saving them.
    Args:
        input_folder: Path to the original dataset folder.
        output_folder: Path to save the processed dataset.
        mean: Mean of the Gaussian noise.
        std: Standard deviation of the Gaussian noise.
    """

    tasks = []
    for class_name in os.listdir(input_folder):
        class_path = os.path.join(input_folder, class_name)
        if os.path.isdir(class_path):
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                if image_path.lower().endswith(('.png', '.jpg', '.jpeg')):
                    output_path = os.path.join(output_folder, class_name, image_name)
                    tasks.append((image_path, output_path, mean, std))

    print(f"Total images to process: {len(tasks)}")

    # Use multiprocessing to process images in parallel with progress bar
    with Pool(processes=cpu_count()) as pool:
        for result in tqdm(pool.imap_unordered(process_image, tasks), total=len(tasks)):
            print(result)


```

```python
if __name__ == "__main__":
    input_folder = base_dir  # Replace with the input folder path
    output_folder = "/content/drive/MyDrive/Food Image Datasets/North Indian/gaussian"    # Replace with the output folder path
    mean = 0      # Mean of Gaussian noise
    std = 25      # Standard deviation of Gaussian noise

    print("Starting the image processing...")
    process_dataset(input_folder, output_folder, mean, std)
    print("Processing complete!")
```

SALT PEPPER NOISE

```python
import os
import cv2
import numpy as np
from multiprocessing import Pool, cpu_count
from tqdm import tqdm  # For progress bar

def apply_salt_pepper_noise(image, salt_prob=0.01, pepper_prob=0.01):
    """
    Apply salt-and-pepper noise to an image.
    Args:
        image: Input image as a NumPy array.
        salt_prob: Probability of salt noise (white pixels).
        pepper_prob: Probability of pepper noise (black pixels).
    Returns:
        Noisy image as a NumPy array.
    """
    noisy_image = image.copy()
    total_pixels = image.shape[0] * image.shape[1]

    # Salt noise (white pixels)
    num_salt = int(total_pixels * salt_prob)
    salt_coords = [np.random.randint(0, i - 1, num_salt) for i in image.shape[:2]]
    noisy_image[salt_coords[0], salt_coords[1]] = 255

    # Pepper noise (black pixels)
    num_pepper = int(total_pixels * pepper_prob)
    pepper_coords = [np.random.randint(0, i - 1, num_pepper) for i in image.shape[:2]]
    noisy_image[pepper_coords[0], pepper_coords[1]] = 0

    return noisy_image

def process_image(args):
    """
    Process a single image by adding salt-and-pepper noise and saving it.
    Args:
        args: A tuple containing image_path, output_path, salt_prob, and pepper_prob.
    """
    image_path, output_path, salt_prob, pepper_prob = args
    try:
        image = cv2.imread(image_path)
        if image is not None:
            noisy_image = apply_salt_pepper_noise(image, salt_prob, pepper_prob)
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            cv2.imwrite(output_path, noisy_image)
            return f"Saved: {output_path}"
        else:
            return f"Failed to read image: {image_path}"
    except Exception as e:
        return f"Error processing {image_path}: {e}"

def process_dataset(input_folder, output_folder, salt_prob=0.01, pepper_prob=0.01):
    """
    Process the dataset by adding salt-and-pepper noise to all images and saving them.
    Args:
        input_folder: Path to the original dataset folder.
        output_folder: Path to save the processed dataset.
        salt_prob: Probability of salt noise (white pixels).
        pepper_prob: Probability of pepper noise (black pixels).
    """
    tasks = []
    for class_name in os.listdir(input_folder):
        class_path = os.path.join(input_folder, class_name)
        if os.path.isdir(class_path):
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                if image_path.lower().endswith(('.png', '.jpg', '.jpeg')):
                    output_path = os.path.join(output_folder, class_name, image_name)
                    tasks.append((image_path, output_path, salt_prob, pepper_prob))

    print(f"Total images to process: {len(tasks)}")

    # Use multiprocessing to process images in parallel with progress bar
    with Pool(processes=cpu_count()) as pool:
        for result in tqdm(pool.imap_unordered(process_image, tasks), total=len(tasks)):
            print(result)

# Example usage
if __name__ == "__main__":
    input_folder = base_dir  # Replace with the input folder path
    output_folder = "/content/drive/MyDrive/Food Image Datasets/North Indian/saltpepper"  # Replace with the output folder path
    salt_prob = 0.01   # Probability of salt noise
    pepper_prob = 0.01  # Probability of pepper noise

    print("Starting the image processing...")
    process_dataset(input_folder, output_folder, salt_prob, pepper_prob)
    print("Processing complete!")
```
